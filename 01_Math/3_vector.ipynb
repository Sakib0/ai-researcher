### 1. What Is a Vector (Conceptually)?

A vector is a mathematical object that has:

Magnitude (how long it is)

Direction (where it points)

Examples of vectors:

Velocity (speed + direction)

Force

Position change (movement from one point to another)

A vector in 2D is written as:

v = [v₁, v₂]
or
v = (x, y)

Example:
v = (3, 4) means:
move 3 units right, 4 units up.

### How Vectors Operate in Python

In Python, vectors are typically represented using:


```python
#Option A: Tuples or Lists
v = [3, 4]
```

```python
#NumPy arrays (most common in math)
import numpy as np
v = np.array([3, 4])
```

### ***Common vector operations:***

```python
#Addition
u = np.array([1, 2])
v = np.array([3, 4])
u + v   # [4, 6]
```

```python
#Scaler Multiplication
2 * v   # [6, 8]
```

```python
#Magnitude
np.linalg.norm(v)   # sqrt(3^2 + 4^2) = 5
```

```python
#Dot Product
np.dot(u, v)   # (1*3 + 2*4) = 11
```

### Visualizing Vectors as Arrows in Python
The best tool for vector visualization is matplotlib, specifically the function:

plt.quiver
This draws arrows originating from any point.

### ***Drawing a Vector from Origin***

```python

import numpy as np
import matplotlib.pyplot as plt

# Define a vector
v = np.array([3, 4])

plt.figure(figsize=(8,6))

# Draw arrow: (start_x, start_y, dx, dy)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')

# Set axis limits
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.gca().set_aspect('equal', adjustable='box')
plt.title("Vector Visualization: v = (3, 4)")
plt.show()
```

### ***Visualizing Multiple Vectors***

```python

u = np.array([2, 1])
v = np.array([3, 4])

plt.figure(figsize=(6,6))

plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='green')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.gca().set_aspect('equal')
plt.title("Two Vectors")
plt.show()
```

### ***Visualizing Vector Addition Geometrically***

You can show that:

u + v = (u₁ + v₁, u₂ + v₂)

```python
u = np.array([2, 1])
v = np.array([3, 4])
w = u + v

plt.figure(figsize=(6,6))

# u
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='green')
# v
plt.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')
# result
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='red')

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.gca().set_aspect('equal')
plt.title("Vector Addition: u + v")
plt.show()
```
