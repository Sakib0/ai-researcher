 ## Simple Math Function in Python 


### ***1.Linear Functions Y=K * X***
A linear function is fundamental in Machine Learning (ML), especially in the context of linear models for regression and classification. It represents the simplest relationship between input features and an output.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Y=2X [The simplest function where multiply input by 2]
x=4
y=x *2
print(f"When x = {x}, y = {y}")
x = 5
y = 2 * x
print(f"When x = {x}, y = {y}")
```

```python
# Plotting the function
x_value=np.linspace(0,10,100)
y_value=x_value *2

plt.figure(figsize=(8,6))
plt.plot(x_value,y_value,'r-',linewidth=2,label='Y=2x')
plt.title(f"Linear Function: Y=2 * X")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Now Try different values of K
K_value=[1,2,0.5,-1,-2]
x=np.linspace(-5,5,100)
plt.figure(figsize=(8,6))
for k in K_value:
    y=k * x
    plt.plot(x,y,label=f'y = {k}x')
plt.title(f"Different Linear Function: Y=K * X")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.grid(True)
plt.tight_layout()
```

### **2. Quadratic Function Y= x**
A quadratic function is a polynomial function of degree two. It is commonly used in ML for modeling non-linear relationships between variables.

```python
# Example: y = x
examples = [-2, -1, 0, 1, 2]
for x in examples:
    y = x ** 2
    print(f"When x = {x}, Y = x = {y}")

# Plot it
x_values = np.linspace(-5, 5, 100)
y_values = x_values ** 2

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, 'g-', linewidth=2, label='y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function: Y = x')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='r', linewidth=0.5)
plt.axvline(x=0, color='r', linewidth=0.5)
plt.show()
```

### ***3. Cubic Functions: y = x***
A cubic function is a polynomial function of degree three. It is used in ML to model more complex non-linear relationships between variables.

```python
# Example: y = x
examples = [-3, -2, -1, 0, 1, 2, 3]
for x in examples:
    y = x ** 3
    print(f"When x = {x}, y = x = {y}")

# Plot it
x_values = np.linspace(-5, 5, 100)
y_values = x_values ** 3

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, 'r-', linewidth=2, label='y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function: y = x')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

### ***4. Square Root Function: y = x***
A square root function is used in ML for data transformation and normalization, especially when dealing with skewed data distributions.

```python
# Example: y = x
examples = [1,4,9,16,25]
for x in examples:
    y = np.sqrt(x)
    print(f"When x = {x}, y = x = {y}")

# Plot it
x_values = np.linspace(0, 25, 100)
y_values = np.sqrt(x_values) 

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, 'b-', linewidth=2, label='y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Square root Function: y = x')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

### ***5. Exponential Function: y = 2^x***
An exponential function is used in ML for modeling growth processes, such as population growth or the spread of information.

```python
# Example: y = 2^x
examples = [0, 1, 2, 3, 4, 5]
for x in examples:
    y = 2 ** x
    print(f"When x = {x}, y = 2^{x} = {y}")
#Plot it
x_values = np.linspace(-2, 5, 100)
y_values = 2 ** x_values

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, 'c-', linewidth=2, label='y = 2^x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Function: y = 2^x')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

### ***6. Sine Function: y = sin(x)***
A sine function is used in ML for modeling periodic phenomena, such as seasonal patterns in time series data.

```python
# Example: y = sin(x)
import math

examples = [0, math.pi/2, math.pi, 3*math.pi/2, 2*math.pi]
for x in examples:
    y = math.sin(x)
    print(f"When x = {x:.2f}, y = sin(x) = {y:.2f}")
# Plot it
x_values = np.linspace(0, 4*np.pi, 100)
y_values = np.sin(x_values)

plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, 'green', linewidth=2, label='y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Function: y = sin(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.show()
```

### ***8. Summary: Compare All Functions***

```python
x = np.linspace(0, 5, 100)

plt.figure(figsize=(12, 8))

# Linear
plt.subplot(2, 2, 1)
plt.plot(x, 2*x, 'b-', linewidth=2)
plt.title('y = 2x (Linear)')
plt.grid(True)

# Quadratic
plt.subplot(2, 2, 2)
plt.plot(x, x**2, 'r-', linewidth=2)
plt.title('y = x (Quadratic)')
plt.grid(True)

# Exponential
plt.subplot(2, 2, 3)
plt.plot(x, 2**x, 'g-', linewidth=2)
plt.title('y = 2^x (Exponential)')
plt.grid(True)

# Square root
plt.subplot(2, 2, 4)
plt.plot(x, np.sqrt(x), 'm-', linewidth=2)
plt.title('y = x (Square Root)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
