### Derivatives in Python 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple numerical derivative function
def numerical_derivative(f, x, h=0.001):
    """Calculate derivative numerically using central difference: (f(x+h) - f(x-h)) / (2h)"""
    return (f(x + h) - f(x - h)) / (2 * h)
```

### ***1. What is a Derivative?***
A derivative measures how quickly a function changes as its input shifts. It captures the functionâ€™s instant rate of change

```python
# Example: y = 2x
# The derivative of y = 2x is always 2
# This means: for every step in x, y increases by 2
def f(x):
    return 2 * x
# The derivative at any point is 2
x_point = 3
derivative_value = 2  # derivative of 2x is 2
print(f"For y = 2x, at x = {x_point}, the derivative (slope) = {derivative_value}")
```

```python
# Visualize function and its derivative
x = np.linspace(0, 10, 100)
y = f(x)
dy_dx = np.full_like(x, 2)  # derivative is constant at 2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot function
ax1.plot(x, y, 'b-', linewidth=2, label='y = 2x')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Function: y = 2x')
ax1.grid(True)
ax1.legend()
# Plot derivative
ax2.plot(x, dy_dx, 'r-', linewidth=2, label="dy/dx = 2")
ax2.set_xlabel('x')
ax2.set_ylabel('dy/dx')
ax2.set_title('Derivative: dy/dx = 2 (constant)')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```

### ***2. Derivative of y = xÂ²***
The derivative of y = xÂ² is y' = 2x. This means the slope changes as x changes!

```python
# Example: y = xÂ², derivative = 2x
def f(x):
    return x*x

def df_dx(x):
    return 2 * x

points = [-3,-2, -1, 0, 1, 2, 3]
print("Point | Function value | Derivative (slope)")
print("-" * 50)

for x in points:
    y=f(x)
    slope=df_dx(x)
    print(f"x={x:3d} | y = {y:3d}        | dy/dx = {slope:3d}")
```

```python
# Visualize
x = np.linspace(-3, 3, 100)
y = f(x)
dy_dx = df_dx(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
# Plot function
ax1.plot(x, y, 'orange', linewidth=2, label='y = xÂ²')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Function: y = xÂ²')
ax1.grid(True)
ax1.legend()
ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)

# Plot function
ax2.plot(x, dy_dx, 'r-', linewidth=2, label='y = 2x')
ax2.set_xlabel('x')
ax2.set_ylabel('dy_dx')
ax2.set_title('Derivative Function: dy_dx = 2x')
ax2.grid(True)
ax2.legend()
ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.axvline(x=0, color='k', linewidth=0.5)
```

### ***3. Visualizing Tangent Lines***
The derivative gives us the slope of the tangent line at any point. Let's draw the tangent lines on the curve!

```python
# Draw tangent lines at different points on y = xÂ²

def f(x): return x**2
def df(x): return 2*x

a = 3.0
x = np.linspace(a-2, a+2, 400)
y = f(x)

tangent = f(a) + df(a)*(x - a)

plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x)=x^2')
plt.plot(x, tangent, '--', label=f'Tangent at x={a:.1f}')
plt.scatter([a], [f(a)], color='black')
plt.legend()
plt.axhline(0, linewidth=0.5); plt.axvline(0, linewidth=0.5)
plt.title('Function and tangent line')
plt.show()
```

### ***4. Common Derivative Rules***
Here are some simple rules for finding derivatives:

```python
# Rule 1: Derivative of constant is 0
# y = 5 â†’ dy/dx = 0

# Rule 2: Derivative of x^n is n*x^(n-1)
# y = xÂ³ â†’ dy/dx = 3xÂ²
# y = xâ´ â†’ dy/dx = 4xÂ³

# Rule 3: Derivative of constant * function
# y = 3xÂ² â†’ dy/dx = 6x

# Let's verify Rule 2 with examples
print("Function    | Derivative")
print("-" * 35)
print("y = xÂ²      | dy/dx = 2x")
print("y = xÂ³      | dy/dx = 3xÂ²")
print("y = xâ´      | dy/dx = 4xÂ³")
print("y = xâµ      | dy/dx = 5xâ´")
print()
print("Let's check y = xÂ³:")

def f_cubic(x):
    return x ** 3

def df_cubic_dx(x):
    return 3 * x ** 2
test_points = [0, 1, 2, 3]
print("\nPoint | Function | Derivative")
print("-" * 35)
for x in test_points:
    y = f_cubic(x)
    slope = df_cubic_dx(x)
    print(f"x={x}   | y = {y:3d}  | dy/dx = {slope:3d}")
```

### ***5. Derivative of y = xÂ³***
Let's visualize the function and its derivative together.

```python
# y = xÂ³, derivative = 3xÂ²
def f(x):
    return x ** 3

def df_dx(x):
    return 3 * x ** 2

x = np.linspace(-2, 2, 100)
y = f(x)
dy_dx = df_dx(x)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Plot function
ax1.plot(x, y, 'g-', linewidth=2, label='y = xÂ³')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Function: y = xÂ³')
ax1.grid(True)
ax1.legend()
ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)
# Plot derivative
ax2.plot(x, dy_dx, 'orange', linewidth=2, label="dy/dx = 3xÂ²")
ax2.set_xlabel('x')
ax2.set_ylabel('dy/dx')
ax2.set_title('Derivative: dy/dx = 3xÂ²')
ax2.grid(True)
ax2.legend()
ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()
```

### ***6. Derivative of y = mx + c***
The derivative of a sum is the sum of derivatives.Ex: y = 2x + 3 â†’ dy/dx = 2 (the constant 3 disappears!)

```python
# y = 2x + 3, derivative = 2
def f(x):
    return 2 * x + 3

def df_dx(x):
    return 2  # derivative of 2x is 2, derivative of 3 is 0

x = np.linspace(-5, 5, 100)
y = f(x)
dy_dx = np.full_like(x, 2)  # constant derivative

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot function
ax1.plot(x, y, 'purple', linewidth=2, label='y = 2x + 3')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Function: y = 2x + 3')
ax1.grid(True)
ax1.legend()
ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)
# Plot derivative
ax2.plot(x, dy_dx, 'r-', linewidth=2, label="dy/dx = 2")
ax2.set_xlabel('x')
ax2.set_ylabel('dy/dx')
ax2.set_title('Derivative: dy/dx = 2 (constant)')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()

print("Note: The constant +3 doesn't affect the derivative!")
print("The slope is always 2, regardless of where we are on the line.")
```
